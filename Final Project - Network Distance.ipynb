```python
# set the working environment 
import pandas as pd
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import networkx as nx
import osmnx as ox, geopandas as gpd
%matplotlib inline
ox.config(log_file=True, log_console=True, use_cache=True)
```

The examples below define the starting point as Union Station.

Change the coordinates to fit your project.

The code saves the output as an image -- you can edit the filename in the "filename" section

```python
# define the lat/lon for your transit station, transit stop, or project site
starting_point = (34.056226, -118.237191)

# calculate the WALKING network within a quarter-mile or 402 meters from the starting point
walking = ox.graph_from_point(starting_point, distance=402, distance_type='bbox', network_type='walk')
### Change the filename below to describe your starting point ###
fig, ax = ox.plot_graph(walking,save=True, file_format='jpg',filename='union_station_walking_quartermile', node_size=30, node_color='#66cc66',show=False,close=False)
# the red dot shows the point we calculated the network distance from
ax.scatter(-118.237191,34.056226, c='red')
```
